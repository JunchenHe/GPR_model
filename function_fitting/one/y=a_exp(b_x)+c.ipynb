```python
import sys
sys.path.append(r'../../')
from get_data import Data

input_dim=(0,)
train_num=52
data=Data("../.././data/data.txt",input_dim,train_num)

#训练指数模型

from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np

def func(x, a, b, c):
    return a * np.exp(b * x) + c

data.train.mean_and_std()
#归一化

xdata=data.train.mean_and_std_X[:,0]
ydata=data.train.mean_and_std_Y[:,0]

popt, pcov = curve_fit(func, xdata, ydata,maxfev=500000)
#popt数组中，三个值分别是待求参数a,b,c
#y2 = [func(i, popt[0],popt[1],popt[2]) for i in xdata]
print("模型参数")
print(popt)

data.test.mean_and_std(data.train.mean_x,data.train.mean_y,
                        data.train.std_x,data.train.std_y)

pre_y=[]
for x in data.test.mean_and_std_X:
    pre_y.append(func(x,popt[0],popt[1],popt[2]))
pre_y=np.array(pre_y)

data.test.re_mean_and_std(pre_y)


err_sum=0.0
print("\n真实值\t预测值\t\t误差率")
for a,b in zip(data.test.Y , data.test.predict_Y):
    err_sum+=abs(a-b)/a
    print(a,b,abs(a-b)/a)
print("平均误差率:",err_sum/data.test.Y.shape[0])
    
```
