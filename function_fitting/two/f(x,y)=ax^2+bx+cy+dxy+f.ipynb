```python
import sys
sys.path.append(r'../../')
from get_data import Data

input_dim=(0,1,)
train_num=52
data=Data("../.././data/data.txt",input_dim,train_num)

#训练指数模型

from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model

def func_change(origin_x):
    a=origin_x[:,0]
    b=origin_x[:,1]

    xdata=[]

    tmp=[]
    for i in range(a.shape[0]):
        tmp.append(a[i]**2)
    xdata.append(np.array(tmp))

    tmp=[]
    for i in range(a.shape[0]):
        tmp.append(a[i])
    xdata.append(np.array(tmp))

    tmp=[]
    for i in range(a.shape[0]):
        tmp.append(b[i])
    xdata.append(np.array(tmp))

    tmp=[]
    for i in range(a.shape[0]):
        tmp.append(b[i]*a[i])
    xdata.append(np.array(tmp))

   
    return np.array(xdata).T

xdata=func_change(data.train.X)
ydata=data.train.Y[:,0]


# 训练数据
regr = linear_model.LinearRegression()
regr.fit(xdata,ydata)
print('coefficients(b1,b2...):',regr.coef_)
print('intercept(b0):',regr.intercept_)

def fit_func(x, a ,b,c,d,e):
    return a*x[0]+b*x[1]+c*x[2]+d*x[3]+e

predict_x=func_change(data.test.X)
pre_y=[]
for x in predict_x:
    pre_y.append([fit_func(x,regr.coef_[0],regr.coef_[1],regr.coef_[2],regr.coef_[3],regr.intercept_)])
pre_y=np.array(pre_y)

err_sum=0.0
print("\n真实值\t预测值\t\t误差率")
for a,b in zip(data.test.Y , pre_y):
    err_sum+=abs(a-b)/a
    print(a,b,abs(a-b)/a)
print("平均误差率:",err_sum/data.test.Y.shape[0])
```
